```python
# just practice


INF = int(1e9)
# 노드와 간선정보
n,m = map(int, input().split())

start = int(input())

graph = [[] for _ in range(n+1)]

d = [INF] * (n + 1)

visited = [False] * (n + 1)

for i in range(m):
    a,b,c = map(int,input().split())
    graph[a].append((b,c))
    
def smallest_index():
    min_value = INF
    index = 0
    for test in range(1, n+1):
        if d[test] < min_value and not visited[test]:
            min_value = d[test]
            index = test
    return index

def dijstra(start):
    d[start] = 0
    visited[start] = True
    
    for i in graph[start]:
        d[i[0]] = i[1]
    
    for i in range(n-1):
        node = smallest_index()
        visited[node] = True
        
        for j in graph[node]:
            cost = d[node] + j[1]            
            if cost < d[j[0]]:
                d[j[0]] = cost


dijstra(start)

for check in range(1, n+1):
    if d[check] == INF:
        print("INF!!")
    else :
        print(d[check])
```

```python
# floid warshall

INF = int(1e9)

#노드와 간선
n = int(input())
m = int(input())

graph = [[INF] * (n+1) for _ in range(n+1)]

for a in range(1, n+1):
    for b in range(1, n+1):
        if a == b:
            graph[a][b] = 0
            
for _ in range(m):
    a,b,c = map(int,input().split())
    
    graph[a][b] = c
    
for k in range(1, n+1):
    for a in range(1, n+1):
        for b in range(1, n+1):
            graph[a][b] = min(graph[a][b], graph[a][k] + graph[k][b])
            
for a in range(1, n+1):
    for b in range(1, n+1):
        if graph[a][b] == INF:
            print("no!! INF!!", end=' ')
        else:
            print(graph[a][b], end=' ')
    print()
```

```python
# 전보
import heapq

INF = int(1e9)

n,m,start = map(int,input().split())

graph = [[] for _ in range(n+1)]

d = [INF] * (n + 1)

for i in range(m):
    a,b,c = map(int,input().split())
    graph[a].append((b,c))

def dijstra(start):
    q = []
    heapq.heappush(q, (0, start))
    d[start] = 0
    
    while q:
        dist, now = heapq.heappop(q)
        
        if d[now] < dist:
            continue
        
        for i in graph[now]:
            cost = i[1] + dist
            if cost < d[i[0]]:
                d[i[0]] = cost
                heapq.heappush(q, (cost, i[0]))

dijstra(start)

count = 0
temp_num = 0
for check in range(1, n+1):
    if d[check] != INF:
        count+=1
        if d[check] > temp_num:
            temp_num = d[check]
# 시작 노드를 제외해야한다... count - 1
print(count-1, temp_num)
```
