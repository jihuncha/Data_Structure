# 바닥 전사

가로 일이가 N / 세로 길이가 2인 타일이 있다.

1x2 / 2x1 / 2x2 타일로 메꾸어야한다.

2x3 의 타일은 메꾸는 수는 5가지

2xn의 타일을 메꾸는 경우의 수 를 모두 구하시오.

1<=N<=1000

#### 출력 조건
796796 으로 나눈 나머지를 출력해라 -> 결과값이굉장히커질수도잇기떄문



### 입력 예시

3

### 출력 예시

5


```python
n = int(input())

d = [0] * 1001

d[1] = 1
d[2] = 3

for i in range(3, n + 1):
    d[i] = (d[i - 1] + 2 * d[i - 2]) % 796796
    
print(d[n])
```
