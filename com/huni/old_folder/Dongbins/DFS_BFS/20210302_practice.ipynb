# 음료수 얼려 먹기
N X M 크기의 얼음틀이 존재

구멍이 뚫려 있다 -> 0

칸막이가 존재 -> 1

구멍이 뚫려있는 부분끼리 상하좌우로 붙어 있는 경우는 연결된 것으로 간주함

엄음틀 모양이 주어졌을때 생성되는 총 아이스큼의 갯수를 구하는 프로그램을 작성해라

#### input
00110

00011

11111

00000

#### output
3

#### 입력 예시 -> 책 잠고
15 14

#### 출력 예시 

8


```python
n,m = map(int,input().split())

result_list = []
for make_result in range(n):
    result_list.append(list(map(int, input())))
    
print(result_list)
    
def dfs(x,y):
    if x < 0 or x >= n or y < 0 or y >= m:
        return False
    
    if result_list[x][y] == 0:
        result_list[x][y] = 1
        dfs(x-1, y)
        dfs(x, y-1)
        dfs(x+1, y)
        dfs(x, y+1)
        return True
    return False

count = 0
for mak_count in range(n):
    for mak_count_second in range(m):
#         print(result_list[0][4])
        if dfs(mak_count, mak_count_second) == True:
            count+=1
print(count)
```

# 미로탈출

N X M 의 직사각형 미로에 갇혀있음.

처음위치는 (1,1) 미로의 출구는 (N,M) 에 위치하며 한번에 한 칸 씩 이동가능하다.

괴물이 있는 부분은 0, 괴물이 있는 부분은 1로 표시

탈출하기 위해 움직여야 하는 최소 칸의 개수를 구하시오.

## 입력 예시

5 6

101010

111111

000001

111111

111111

## 출력 예시

10

```python
from collections import deque

n,m = map(int,input().split())

graph = []
for make_result in range(n):
    graph.append(list(map(int, input())))
    
print(graph)

dx = [0,0,1,-1]
dy = [1,-1,0,0]

def bfs(x,y):
    queue = deque()
    queue.append((x,y))
    
    while queue:
        x, y = queue.popleft()
        
        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]
            
            if nx < 0 or nx >= n or ny < 0 or ny >= m:
                continue
            
            if graph[nx][ny] == 0:
                continue
                
            if graph[nx][ny] == 1:
                graph[nx][ny] = graph[x][y] + 1
                queue.append((nx,ny))
            
        print(queue)
    
    return graph[n -1][m - 1]

print(bfs(0,0))
print(graph)    
```
