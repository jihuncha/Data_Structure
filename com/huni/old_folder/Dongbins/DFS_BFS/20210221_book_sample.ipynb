```python
from collections import deque

queue = deque()

queue.append(5)
queue.append(2)
queue.append(3)
queue.append(7)
queue.popleft()
queue.append(1)
queue.append(4)
queue.popleft()

print(queue)
queue.reverse()
print(queue)





```

```python
def recursive_function(i):
    if i == 100:
        return
    print(i, " 번 째 재귀 함수에서 ", i+1 , " 번 째 재귀 함수를 호출합니다.")
    recursive_function(i + 1)
    print(i, " 번 째 재귀 함수를 종룧랍니다.")

recursive_function(1)
```

```python
#반복적으로 구현한 factorial

def factorial_iterative(n):
    result = 1
    for mak_fact in range(1, n + 1):
        result *= mak_fact
    return result

print(factorial_iterative(5))

#재귀적으로 구현한 factorial
def factorial_recursive(n):
    if n <= 1:
        return 1;
    else:
        return n * factorial_recursive(n-1)

print(factorial_recursive(5))
```
